```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = pd.read_csv("C:\\Users\\khyat\\Downloads\\IRIS.csv")

```

```python
# Assuming you have a dataset stored in a DataFrame called 'data'
data = pd.DataFrame({'x': [1, 2, 3, 10, 11, 13],
                     'y': [10, 12, 13, 1, 3, 2]})

# Creating an array of the dataset values
X = np.array(data)

# Initializing the K-means clustering algorithm with 2 clusters
kmeans = KMeans(n_clusters=2)

# Training the K-means model on the data
kmeans.fit(X)

# Getting the cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='red', s=150)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-means Clustering')
plt.show()
```
